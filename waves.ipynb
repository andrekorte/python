## Phase Shift

```python
t = np.linspace(0, 2*np.pi, 100)
f  = 1.0

def phase_shift(phi=np.pi):
    fig = plt.figure(figsize=(9, 4))
    plt.xlim(0, 2*np.pi)
    plt.plot(t, np.sin(t), '.', lw=4, label='sin(t)')
    plt.plot(t, np.cos(t), '.', lw=4, label='cos(t)')
    plt.plot(t, np.sin(t + phi), lw=2)
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.legend()
             
v = interactive(phase_shift, phi=(0, 2*np.pi, 0.01))
display(v)
```

## Square and Sawtooth Waves from Sine Waves

```python
def squarewave(t, f, k):
    x = np.zeros(np.size(t))
    for k in xrange(1, k+1, 2):
        x = x + np.sin(2*np.pi*k*f*t)/k
    return x
    
def sawtooth(t, f, k):
    x = np.zeros(np.size(t))
    for k in xrange(1, k+1):
        x = x + np.sin(2*np.pi*k*f*t)/k
    return x

def plot_both(k):
    fig, axes = plt.subplots(1, 2, figsize=(16,4))
    
    
    axes[0].set_xlabel('t [s]')
    axes[0].set_ylabel('y(t)')
    axes[0].plot(t, squarewave(t, f, k), lw=2)
    
    axes[1].set_xlabel('t [s]')
    axes[1].set_ylabel('y(t)')
    axes[1].plot(t, sawtooth(t, f, k), lw=2)
```

```python
t = np.linspace(0, 1, 512)
f = 2.0
v = interactive(plot_both, k=(1, 15))
display(v)
```

### The Gibbs' effect

```python
def gibbs_plot():    
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))
    suptitle = fig.suptitle("The building of a square wave: Gibbs' effect", fontsize=20)

    axes[0].set_xlabel('t')
    axes[0].set_ylabel('y(t)')

    axes[1].set_xlim([0, 0.5])
    axes[1].set_ylim([0, 1])
    axes[1].set_xlabel('t')
    axes[1].set_ylabel('y(t)')

    for k in xrange(1, max_k+1, 2):
        if k in [1, 9, 19, 25]:
            axes[0].plot(t, squarewave(t, f, k), label='k = {0}'.format(k), linewidth=2.0)
            axes[1].plot(t, squarewave(t, f, k), label='k = {0}'.format(k), linewidth=2.0)
    
    axes[0].plot(t, squarewave(t, f, 75), label='k = 75'.format(k), linewidth=2.0)
    axes[0].legend()

    axes[1].plot(t, squarewave(t, f, 75), label='k = 75'.format(k), linewidth=2.0)
    axes[1].legend(loc=8)

t = np.linspace(0, 1, 500)
f = 1
max_k = 25
gibbs_plot()
```

## Beat Frequencies

```python
def beat_freq(f1=220.0, f2=224.0):
    max_time = 3
    rate = 8000
    t = np.linspace(0,max_time,rate*max_time)
    signal = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2*t)
    plt.plot(t, signal)
    plt.xlabel('t')
    plt.ylabel('y(t)')
    print(f1, f2, abs(f1-f2))
    display(Audio(data=signal, rate=rate))
    
v = interactive(beat_freq, f1=(200.0,300.0), f2=(200.0,300.0))
display(v)
```

## Fourier Transform

- Fourier transformation of a pure sine with no phase shift will result in pure imaginary signal.
- Fourier transformation of a pure cosine with no phase shift will result in pure real signal.

- check out the inverse transform
- 20*np.log10(FFT) Power spectrum?

```python
from scipy import signal

# Signal is a square wave sampled 512 times on an interval of one second
t = np.linspace(0, 1, 512)
sig = signal.square(2 * np.pi * 5 * t)

# Real Fourier Transform
FFT = abs(np.fft.rfft(sig))
freqs = np.fft.rfftfreq(sig.size, t[1]-t[0])
db = 20*np.log10(FFT)

def plot_fft_square():
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))
    fig.suptitle('Discrete Fourier Transform of a square wave', y=0)
    axes[0].plot(t, sig, lw=2, label='signal')
    axes[0].set_ylim(-2, 2)
    axes[0].set_xlabel('t [s]')
    axes[0].set_ylabel('y(t)')
    axes[0].legend()

    axes[1].plot(freqs, FFT, lw=2, label='DFT')
    axes[1].set_xlim(0, len(t)/2)
    axes[1].set_xlabel('f [Hz]')
    axes[1].set_ylabel('FT')
    axes[1].legend()
    
plot_fft_square()
```

### Power Spectral Density

```python
# Generate a test signal, a 2 Vrms sine wave at 1234 Hz, corrupted by 0.001 V**2/Hz of white noise sampled at 10 kHz.

fs = 10e3
N = 1e5
amp = 2*np.sqrt(2)
freq = 1234.0
noise_power = 0.001 * fs / 2
time = np.arange(N) / fs
x = amp*np.sin(2*np.pi*freq*time)
x += np.random.normal(scale=np.sqrt(noise_power), size=time.shape)

# Compute and plot the power spectral density.

fig, axes = plt.subplots(1, 2, figsize=(16, 4))

f, Pxx_den = signal.periodogram(x, fs)
axes[0].semilogy(f, Pxx_den, lw=2)
axes[0].set_ylim([1e-7, 1e2])
axes[0].set_xlabel('frequency [Hz]')
axes[0].set_ylabel('PSD [V**2/Hz]')

# If we average the last half of the spectral density, to exclude the peak, we can recover the noise power on the signal.

#print(np.mean(Pxx_den[256:]))
#0.0009924865443739191

# Now compute and plot the power spectrum.

f, Pxx_spec = signal.periodogram(x, fs, 'flattop', scaling='spectrum')
axes[1].semilogy(f, np.sqrt(Pxx_spec), lw=2)
axes[1].set_ylim([1e-4, 1e1])
axes[1].set_xlabel('frequency [Hz]')
axes[1].set_ylabel('Linear spectrum [V RMS]')

# The peak height in the power spectrum is an estimate of the RMS amplitude.

print('Estimated RMS amplitude: {0} V'.format(np.sqrt(Pxx_spec.max())))
```

## Peakdetection

```python
# There are more methods in this module
import peakdetect

# Signal is a square wave sampled 512 times on an interval of one second with harmonics up to k
t = np.linspace(0, 1, 512)
f = 10
k = 75
sig = squarewave(t, f, k)

# Real Fourier Transform
FFT = abs(np.fft.rfft(sig))
freqs = np.fft.rfftfreq(sig.size, t[1]-t[0])
db = 20*np.log10(FFT)

def plot_peakdetect(lookahead=115):
    _max, _min = peakdetect.peakdetect(FFT, freqs, lookahead, delta=0.30)
    xm = [p[0] for p in _max]
    ym = [p[1] for p in _max]
    fig, axes = plt.subplots(1, 1, figsize=(16, 4))
    axes.plot(freqs, FFT, '-', linewidth=2.0, label='DFT')
    axes.set_xlim([freqs[0], freqs[-1]])
    axes.set_xlabel('f [Hz]')
    axes.set_ylabel('FT')

    axes.plot(xm, ym, 'ro', label='Peaks')
    axes.legend()
    
v = interactive(plot_peakdetect, lookahead=(1, 500))
display(v)
```

```python
def threshold(n=10):
    thresh = FFT.mean() + n*np.std(FFT)
    y = np.array( np.where( FFT >= thresh ))
    ym = [FFT[i] for i in y]
    xm = freqs[y]
    
    fig, axes = plt.subplots(1, 1, figsize=(16, 4))
    axes.plot(freqs, FFT, '-', linewidth=2.0, label='DFT')
    axes.set_xlim([freqs[0], freqs[-1]])
    axes.set_xlabel('f [Hz]')
    axes.set_ylabel('FT')
    axes.plot(xm, ym, 'ro')
    plt.hlines(thresh, freqs[0], freqs[-1], linewidth=2, linestyles='dotted', colors='r', label='Threshold: {0} sigma'.format(n))
    axes.legend(loc=0)
    print('Threshold:{0}'.format(thresh))
    
v = interactive(threshold, n=(0, 25))
display(v)
```

```python
def rel_extrema():
    extrema = signal.argrelextrema(FFT, np.greater)
    x = extrema
    y = [FFT[i] for i in x]
    
    fig, axes = plt.subplots(1, 1, figsize=(16, 4))
    axes.plot(freqs, FFT, '-', linewidth=2.0, label='DFT')
    axes.set_xlim([freqs[0], freqs[-1]])
    axes.set_xlabel('f [Hz]')
    axes.set_ylabel('FT')
    axes.plot(x, y, 'ro')
    axes.legend()

rel_extrema()
```

```python
def plot_cwt(width_upper=5):
    width = np.arange(1, width_upper)
    peakind = signal.find_peaks_cwt(FFT, width)
    
    fig, axes = plt.subplots(1, 1, figsize=(16, 4))
    axes.plot(freqs, FFT, '-', linewidth=2.0, label='DFT')
    axes.set_xlim([freqs[0], freqs[-1]])
    axes.set_xlabel('f [Hz]')
    axes.set_ylabel('FT')
    axes.plot(freqs[peakind], FFT[peakind], 'ro')
    axes.legend()
    
v = interactive(plot_cwt, width_upper=(2, 100))
display(v)
```
